## Drop Track with Null Track Name or Audio Features

```python
df_final = df.dropna(subset = ['Danceability', 'Track Name'])
df_final.isnull().sum()
```

## Impute Missing Record Label

```python
df_final[df_final['Label'].isnull()].head()
```

```python

# Impute missing values for 'Eiffel 65'
df_final.loc[df_final['Artist Name(s)'] == 'Eiffel 65', 'Label'] = df_final.loc[df_final['Artist Name(s)'] == 'Eiffel 65', 'Label'].fillna('Universal Republic Records')

# Impute missing values for 'The White Stripes'
df_final.loc[df_final['Artist Name(s)'] == 'The White Stripes', 'Label'] = df_final.loc[
    df_final['Artist Name(s)'] == 'The White Stripes', 'Label'
].fillna('V2 Records')
```

```python
df_final.isnull().sum()
```

```python
df_final[df_final['Copyrights'].isnull()].head()
```

## Fill Missing Values of Artist Genres and Copyrights

```python
df_final['Artist Genres'].fillna('missing genres', inplace = True)
df_final['Copyrights'].fillna('missing copyrights', inplace = True)
df_final['Track Preview URL'].fillna('missing track preview url', inplace = True)
```

```python
df_final.isnull().sum()
```

## Add Lyrics

```python
import re

df_final['track_name_join'] = df_final['Track Name'].str.lower().str.strip()
df_final['artist_name_join'] = df_final['Artist Name(s)'].str.lower().str.strip()

# lyrics['song_join'] = lyrics['song'].str.lower().str.strip()
# lyrics['artist_join'] = lyrics['artist'].str.lower().str.strip()

# song_lyrics = lyrics[['song_join', 'artist_join','text']].drop_duplicates()

# final = df_final.merge(song_lyrics, left_on = ['track_name_join', 'artist_name_join'], right_on = ['song_join', 'artist_join'], how = 'left')

# final.drop(columns = ['song_join', 'artist_join'], inplace = True)

# Function to extract artist name based on condition
def extract_artist(artist):
    if re.match(r"\[.*\]", artist):  # Check if it contains square brackets
        return re.sub(r"[\[\]']", "", artist)  # Remove brackets and quotes
    return artist

lyrics_2['artists'] = lyrics_2['artists'].astype(str)
lyrics_2['name'] = lyrics_2['name'].astype(str)
lyrics_2['lyrics'] = lyrics_2['lyrics'].astype(str)


# Apply function to extract artist names
lyrics_2["artist_name"] = lyrics_2["artists"].apply(extract_artist)

lyrics_2['song_name_join'] = lyrics_2['name'].str.lower()
lyrics_2['artists_name_join'] = lyrics_2['artist_name'].str.lower()


lyrics_2_clean = lyrics_2[['song_name_join', 'artists_name_join', 'lyrics']]
lyrics_2_unique = lyrics_2_clean.drop_duplicates(subset=['song_name_join', 'artists_name_join'], keep = 'first')

# lyrics_2_clean.reset_index()
final = df_final.merge(lyrics_2_unique, left_on = ['track_name_join', 'artist_name_join'], right_on = ['song_name_join', 'artists_name_join'], how = 'left')

final.drop(columns= ['track_name_join', 'artist_name_join', 'song_name_join', 'artists_name_join', 'Album Genres'], inplace = True)
```
